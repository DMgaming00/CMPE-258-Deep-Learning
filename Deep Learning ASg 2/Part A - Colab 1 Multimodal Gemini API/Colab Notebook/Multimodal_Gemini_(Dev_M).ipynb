## 2Ô∏è‚É£ Secure API Configuration

```python

from google import genai
import os
from getpass import getpass

os.environ["GOOGLE_API_KEY"] = getpass("Enter your Gemini API Key: ")

client = genai.Client(api_key=os.environ["GOOGLE_API_KEY"])

print("‚úÖ API Connected Successfully")
```

## 3Ô∏è‚É£ List Available Models

```python

models = client.models.list()

for m in models:
    print(m.name)
```

## üß† 4Ô∏è‚É£ Text-to-Text Generation

```python

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Create a futuristic conversation between a human and an AI robot living on Mars in 2080."
)

print(response.text)
```

## üìä 5Ô∏è‚É£ Structured Output Example

```python

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="""
Analyze the future of AI in healthcare.

Return output in this format:

1. Key Trends
2. Risks
3. Opportunities
4. Final Summary
"""
)

print(response.text)
```


## üñºÔ∏è 6Ô∏è‚É£ Multimodal Image Understanding

Upload an image when prompted.


```python

from google.colab import files
from PIL import Image
import matplotlib.pyplot as plt

uploaded = files.upload()

image_path = list(uploaded.keys())[0]

img = Image.open(image_path)
plt.imshow(img)
plt.axis("off")

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=[
        "Analyze this image deeply. Describe objects, emotions, setting, and possible story:",
        img
    ]
)

print("\nAI Analysis:\n")
print(response.text)
```
